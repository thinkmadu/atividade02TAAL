```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
AVLInsertion = [20.4587, 45.8324, 86.9958]
RBInsertion = [16.3618, 45.7421, 65.91]
entradas = [30000, 60000, 100000]


plt.figure(figsize=(10, 6))
plt.title('Insero em rvores AVL e Rubro Negra')
plt.xlabel('Tamanho da Entrada')
plt.ylabel('Tempo Mdio (ms)')
plt.grid(True, which='both', linestyle='--', lw=0.5)

plt.plot(entradas, AVLInsertion, label='rvore AVL', color="black")
plt.plot(entradas, RBInsertion, label='rvore Rubro-Negro', color="red")
plt.legend()
plt.show()
```

```python
AVLRemove = [31.0917, 53.0466, 94.1057]
RBRemove = [15.34, 34.5295, 81.1658]
entradas = [30000, 60000, 100000]


plt.figure(figsize=(10, 6))
plt.title('Remoo de item em rvores AVL e Rubro Negra')
plt.xlabel('Tamanho da Entrada')
plt.ylabel('Tempo Mdio (ms)')
plt.grid(True, which='both', linestyle='--', lw=0.5)

plt.plot(entradas, AVLRemove, label='rvore AVL', color="black")
plt.plot(entradas, RBRemove, label='rvore Rubro-Negro', color="red")
plt.legend()
plt.show()
```
